```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import LinearSegmentedColormap
```

```python
contenido_vacunas = pd.read_csv("test_set_features.csv")
```

```python
contenido_vacunas.opinion_h1n1_vacc_effective = contenido_vacunas.opinion_h1n1_vacc_effective.fillna(0)
```

```python
matriz = contenido_vacunas.pivot_table(index="age_group", columns="opinion_h1n1_vacc_effective", values="respondent_id", aggfunc="count")
```

```python
matriz
```

```python
plt.figure(dpi=125, figsize=(7, 6))
fig = plt.figure()
fig.patch.set_facecolor('grey')
sns.heatmap(matriz, cmap = 'Blues_r', annot=True, fmt='g', square=True, linewidth=1, linecolor='black', cbar_kws={"shrink": 0.8}, )
plt.title("Opinion de las encuestas de la efectividad de la vacuna H1N1 por edades")
plt.xlabel("Puntaje de las encuesta (0 no respondio)")
plt.ylabel("Rango de edad")
```

Aclaracion: Considere que aquellas personas que realizaron la encuenta y no dieron un puntaje es porque decidieron no opinar de la efectividad de la vacuna, por eso remplace el valor de los Nan por un 0.
