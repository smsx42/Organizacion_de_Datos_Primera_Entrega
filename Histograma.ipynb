```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
contenido_dengue = pd.read_csv("dengue_labels_train.csv")
```

```python
agrupados_semanas = contenido_dengue.groupby(['weekofyear'])
```

```python
agrupados_semanas = agrupados_semanas.agg({"total_cases":"sum"})
```

```python
plt.figure(dpi=100, figsize=(7, 4.6))
plt.hist(contenido_dengue.weekofyear,bins=11, edgecolor = 'black', weights=contenido_dengue.total_cases, color = '#db001a')
plt.grid(False)
plt.title("Contagios de dengue en las regiones de Iquitos y San Jose en funcion de las semanas entre 1990 y 2010",  fontsize = 10)
plt.xlabel("Semana")
plt.xlim(0, 55)
plt.xticks((1,5,10,15,20,25,30,35,40,45,50,55))
plt.ylabel("Cantidad de contagios")
```
