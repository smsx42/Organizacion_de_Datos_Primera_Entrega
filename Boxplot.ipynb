```python
import seaborn as sns
import matplotlib as mpl
import pandas as pd
from datetime import date
import matplotlib.pyplot as plt
```

```python
datos = pd.read_csv("datasetRofex4 - datasetRofex2.csv")
```

```python
datos.Fecha = datos.Fecha.map(lambda fila: int(fila[6:10]))
```

```python
plt.figure(dpi=110, figsize=(8, 7))
fig = plt.figure()
fig.patch.set_facecolor('grey')
params = {'xtick.labelsize': 10, 'ytick.labelsize': 10}
sns.boxplot(data= datos, x = "Fecha", y = "Cierre", color = 'Green')
plt.xlabel("AÃ±os")
plt.ylabel("Precio de cierre")
plt.title("Datos estadisticos del precio de cierre de la soja durante el periodo 2004 - 2020", fontsize = 11)
plt.xticks(rotation = 90)
mpl.rcParams.update(params)
```
